```python
import matplotlib.pyplot as plt
def lcg(modulus, a, c, seed):
    ls = []
    for i in range(100):
        seed = (a*seed + c) % modulus
        ls.append(seed)
    return ls

rand_ls = lcg(100, 2.5, 33, 1) #调用函数
bins = range(0, 101, 10) #设置分割段数，在1-101中分为10段
plt.hist(rand_ls, bins = bins, color = 'dodgerblue', histtype = 'bar',  alpha = 1) #绘制统计图，设置颜色为蓝色，透明度为0.95
plt.xticks(range(0, 101, 10), fontproperties = 'Times New Roman', size = 15) #设置x轴标签，字体和字体大小
plt.yticks(fontproperties = 'Times New Roman', size = 15) #设置y轴标签，字体和字体大小
plt.xlim(0, 100) #设置x轴范围
plt.xlabel('Numbers', fontproperties = 'Times New Roman', size = 14) #设置x轴标签，字体和字体大小
plt.ylabel('N of each number', fontproperties = 'Times New Roman', size = 14) #设置y轴标签，字体和字体大小
plt.title('Check Uniform Distribution of 100 iterations', fontproperties = 'Times New Roman', size = 14) #设置标题，字体和字体大小
plt.savefig(r'D:\教学\家教\咸鱼\100.jpg', dpi = 100) #输出图片，分辨率为100dpi
plt.show()

```

```python
import matplotlib.pyplot as plt
def lcg(modulus, a, c, seed):
    ls = []
    for i in range(100):
        seed = (a*seed + c) % modulus
        ls.append(seed)
    return ls

rand_ls = lcg(100, 2.5, 33, 1) #调用函数
bins = range(0, 101, 10) #设置分割段数，在1-101中分为10段
plt.hist(rand_ls, bins = bins, color = 'dodgerblue', histtype = 'bar',  alpha = 1) #绘制统计图，设置颜色为蓝色，透明度为0.95

plt.title('Check Uniform Distribution of 100 iterations', fontproperties = 'Times New Roman', size = 14)
plt.xlabel('Numbers', fontproperties = 'Times New Roman', size = 14)
plt.ylabel('N of each number', fontproperties = 'Times New Roman', size = 14)
plt.show()
```
